Amazon Sales Data


Description:

This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details.



Features:

**product_id**: Unique identifier for each product

**product_name**: Name of the product

**category**: Category of the product

**discounted_price**: Discounted price of the product

**actual_price**: Actual price of the product

**discount_percentage**: Percentage of discount for the product

**rating**: Rating of the product (1-5)

**rating_count**: Number of people who voted for the Amazon rating

**about_product**: Description about the product

**user_id**: ID of the user who wrote the review

**user_name**: Name of the user who wrote the review

**review_id**: ID of the user review

**review_title**: Short review

**review_content**: Long review

**img_link**: Image link of the product

**product_link**: Official website link of the product




```python
# importing needed modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skimpy import skim
import warnings 

warnings.filterwarnings('ignore')
```

```python
# get dataset into a Main dataframe by using pandas

main_df=pd.read_csv('amazon.csv')
```

```python
# now make copy of main df for frature operations 

df=main_df.copy()
```

```python
# first 5 rows of df
df.head()
```

```python
# check all details of df
skim(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['discounted_price']=df['discounted_price'].str.replace('₹','')
```

```python
df['discounted_price']=df['discounted_price'].str.replace(',','')
```

```python
df['discounted_price']=df['discounted_price'].astype(float)
```

```python
df.info()
```

```python
df['actual_price']=df['actual_price'].str.replace('₹','')
```

```python
df['actual_price']=df['actual_price'].str.replace(',','')
```

```python
df['actual_price']=df['actual_price'].astype(float)
```

```python
df.info()
```

```python
df['discount_percentage']=df['discount_percentage'].str.replace('%','')
```

```python
df['discount_percentage']=df['discount_percentage'].astype(int)
```

```python
df.info()
```

```python
df['rating']=df['rating'].str.replace('|','0')
```

```python
df['rating']=df['rating'].astype(float)
```

```python
df['rating']
```

```python
df['rating'].unique()
```

```python
df.info()
```

```python
df['rating_count']=df['rating_count'].str.replace(',','')
```

```python
df['rating_count']=df['rating_count'].astype(int)
```

```python
#1. top 5 selling category on amazon with count ?
df['category'].value_counts().head(5)
```

```python
# 2.What are the top 5 most expensive products based on the actual price?

top_5_expensive = df.sort_values(by='actual_price', ascending=False).head(5)
top_5_expensive[['product_name', 'actual_price']]
```

```python
# 3.Which category has the highest average discount percentage?
df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).head(1)
```

```python
# 4.What is the correlation between the discounted price and the actual price?
df[['discounted_price', 'actual_price']].corr()
```

```python
#5.What is the average rating for products in top 5 category?
df.groupby('category')['rating'].mean().sort_values(ascending=False).head(5)
```

```python
#6.How many unique users have written reviews, and what is the distribution of the number of reviews per user?
unique_users = df['user_id'].nunique()
reviews_per_user = df.groupby('user_id').size()
reviews_per_user.describe()
```

```python
#7.What is the distribution of product ratings?
sns.histplot(df['rating'], kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#8. Which product has the highest number of reviews?
df.groupby('product_name').size().idxmax()
```

```python
#9. What is the percentage of products that have a discount percentage greater than 50%?
discount_greater_than_50 = df[df['discount_percentage'] > 50].shape[0]
total_products = df.shape[0]
percentage = (discount_greater_than_50 / total_products) * 100
percentage
```

```python
#10. What is the average rating for products with a discount of more than 30%?
df[df['discount_percentage'] > 30]['rating'].mean()
```

```python
#11. What is the distribution of the actual prices of the products?
plt.figure(figsize=(10, 6))
sns.histplot(df['actual_price'], bins=10, kde=True)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#12. How do product ratings vary across top 5 categories?
top_5_cat = df['category'].value_counts().head(5).index
df_top_5_cat = df[df['category'].isin(top_5_cat)]
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='rating', data=df_top_5_cat)
plt.title('Ratings Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
#13. What is the relationship between the discounted price and the actual price?
plt.figure(figsize=(8, 6))
sns.scatterplot(x='actual_price', y='discounted_price', data=df)
plt.title('Discounted Price vs Actual Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
#14.What is the count of reviews for each product rating?
plt.figure(figsize=(8, 6))
sns.countplot(x='rating', data=df, palette='Blues')
plt.title('Count of Products for Each Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#15.How does the average discount percentage vary across different product categories?
plt.figure(figsize=(12, 6))
avg_discount_per_category = df_top_5_cat.groupby('category')['discount_percentage'].mean()
sns.barplot(x=avg_discount_per_category.index, y=avg_discount_per_category.values, palette='viridis')
plt.title('Average Discount Percentage by Category')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=90)
plt.show()
```
