Title: Car Sale Dataset


Source:

The dataset is available on Kaggle, representing data commonly seen in online car resale platforms.


Description:

The dataset contains details about used cars listed for sale, such as brand, model, selling price, kilometers
driven, fuel type, and transmission type. This information is valuable for predictive modeling, market analysis,
and understanding customer preferences.

Features:

* **Car_id**: A unique identifier for each car in the dataset, helping to track individual car entries.
* **Date**: The date when the car sale transaction took place, formatted as YYYY-MM-DD.
* **Customer Name**: The name of the customer who purchased the car, represented as a string.
* **Gender**: The gender of the customer, categorized as "Male" or "Female.
* **Annual Income**: The customer's annual income in US dollars, represented as a numeric value.
* **Dealer_Name**: The name of the dealership selling the car, represented as a string.
* **Company**: The manufacturer or brand name of the car, such as "Toyota," "Ford," etc.
* **Model**: The specific model name of the car, such as "Corolla," "Civic," etc.
* **Engine**: The engine type of the car, such as "V6," "I4," etc.
* **Transmission**: The type of transmission in the car, either "Manual" or "Automatic.
* **Color**: The color of the car, represented as a string (e.g., "Red," "Blue").
* **Price** ($): The selling price of the car in US dollars.
* **Dealer_No**: A unique identifier for each car dealer in the dataset.
* **Body Style**: The body style of the car, such as "Sedan," "SUV," etc.
* **Phone**: The phone number of the customer who purchased the car.
* **Dealer_Region**: The geographical region of the car dealer, such as "North," "South," etc.

```python
# importing needed modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from skimpy import skim
import warnings 

warnings.filterwarnings('ignore')
```

```python
# get dataset into a Main dataframe by using pandas

main_df=pd.read_csv("Car Sale.csv")
```

```python
# now make copy of main df for frature operations 

df=main_df.copy()
```

```python
# first 5 rows of df
df.head()
```

```python
# check all details of df
skim(df)
```

```python
# here only one null value for Customer name
```

```python
# 1.What is the average selling price of cars for each dealer, and how does it compare across different dealers ?

average_price_by_dealer=df.groupby('Dealer_Name')['Price ($)'].mean()
average_price_by_dealer
```

```python
plt.figure(figsize=(10, 6))
average_price_by_dealer.plot(kind='bar', color='skyblue')
plt.title('Average Selling Price of Cars by Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends ?

price_variation_by_company=df.groupby('Company')['Price ($)'].std()
price_variation_by_company
```

```python
plt.figure(figsize=(10, 6))
price_variation_by_company.plot(kind='bar', color='lightcoral')
plt.title('Price Variation (Standard Deviation) by Car Brand')
plt.xlabel('Car Brand (Company)')
plt.ylabel('Standard Deviation of Price ($)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare ?

# 1. Calculate the Interquartile Range (IQR) for each Transmission type
iqr_by_transmission = df.groupby('Transmission')['Price ($)'].quantile(0.75) - df.groupby('Transmission')['Price ($)'].quantile(0.25)

# 2. Display the IQR for each transmission type
print("IQR for each transmission type:")
print(iqr_by_transmission)

# 3. Visualize the distribution using boxplots
plt.figure(figsize=(10, 6))
df.boxplot(column='Price ($)', by='Transmission', patch_artist=True, showmeans=True, 
           boxprops=dict(facecolor='lightblue', color='blue'), 
           flierprops=dict(markerfacecolor='red', marker='o', markersize=5))
plt.title('Price Distribution by Transmission Type')
plt.suptitle('')  # Removes the extra title generated by boxplot
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
#4.What is the distribution of car prices across different regions ?

plt.figure(figsize=(10, 6))
df.boxplot(column='Price ($)', by='Dealer_Region', patch_artist=True, showmeans=True,
           boxprops=dict(facecolor='lightgreen', color='green'),
           flierprops=dict(markerfacecolor='red', marker='o', markersize=5))
plt.title('Price Distribution by Dealer Region')
plt.suptitle('')  # Removes the extra title generated by boxplot
plt.xlabel('Dealer Region')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. What is the distribution of cars based on body styles ?

body_style_counts = df['Body Style'].value_counts()

# pie chart 
explode = [0.1] + [0] * (len(body_style_counts) - 1)  # Explode the first slice for emphasis

plt.figure(figsize=(8, 8))
plt.pie(body_style_counts, labels=body_style_counts.index, autopct='%1.1f%%', startangle=140, 
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], explode=explode)
plt.title('Distribution of Cars by Body Style')
plt.show()
```

```python
# 6. How does the average selling price of cars vary by customer gender and annual income ?
plt.figure(figsize=(10,5))
sns.barplot(x='Gender',y='Price ($)',data=df,palette='Set2')
plt.title("selling price in Gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x='Annual Income',y='Price ($)',data=df,palette='Set2')
plt.title("selling price in Annual Income ")
plt.show()
```

```python
# 7. What is the distribution of car prices by region, and how does the number of cars sold vary by region ?

plt.figure(figsize=(10, 6))
sns.boxplot(x='Dealer_Region', y='Price ($)', data=df, palette='Set2')
plt.title('Distribution of Car Prices by Region', fontsize=14)
plt.xlabel('Dealer Region', fontsize=12)
plt.ylabel('Car Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
region_counts = df['Dealer_Region'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=region_counts.index, y=region_counts.values, palette='Set3')
plt.title('Number of Cars Sold by Region', fontsize=14)
plt.xlabel('Dealer Region', fontsize=12)
plt.ylabel('Number of Cars Sold', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 8.  How does the average car price differ between cars with different engine sizes ?

df.groupby('Engine')['Price ($)'].mean()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Engine',y='Price ($)',data=df,palette='Set2')
```

```python
# 9. How do car prices vary based on the customer’s annual income bracket ?

income_brackets = pd.cut(df['Annual Income'], bins=10)

df['Income Bracket'] = income_brackets

avg_price_by_income = df.groupby('Income Bracket')['Price ($)'].mean()


plt.figure(figsize=(10, 6))
avg_price_by_income.plot(kind='bar', color='skyblue', title='Average Car Price by Income Bracket')
plt.ylabel('Average Price ($)', fontsize=12)
plt.xlabel('Income Bracket', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 10.What are the top 5 car models with the highest number of sales, and how does their price distribution look ?

sales_count_by_model = df['Model'].value_counts()
top_5_models = sales_count_by_model.head(5).index
df_top_5_models = df[df['Model'].isin(top_5_models)]

#  Visualize boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Model', y='Price ($)', data=df_top_5_models, palette='Set2')
plt.title('Price Distribution of Top 5 Car Models by Sales', fontsize=14)
plt.xlabel('Car Model', fontsize=12)
plt.ylabel('Car Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 11. How does car price vary with engine size across different car colors, and which colors have the highest price variation ?

# Scatter plot: Car price vs. engine size, colored by car color
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine', y='Price ($)', hue='Color', data=df, palette='Set2')
plt.title('Car Price vs. Engine Size across Car Colors', fontsize=14)
plt.xlabel('Engine Size', fontsize=12)
plt.ylabel('Car Price ($)', fontsize=12)
plt.tight_layout()
plt.show()

# Boxplot: Price variation across different car colors
plt.figure(figsize=(10, 6))
sns.boxplot(x='Color', y='Price ($)', data=df, palette='Set3')
plt.title('Price Variation across Car Colors', fontsize=14)
plt.xlabel('Car Color', fontsize=12)
plt.ylabel('Car Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 12.Is there any seasonal trend in car sales based on the date of sale?
# Convert 'Date' to datetime format if not already
df['Date'] = pd.to_datetime(df['Date'])

# Extract month or quarter from the date
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.to_period('Q')

# Bar plot for monthly car sales
plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=df, palette='Set1')
plt.title('Monthly Car Sales', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Cars Sold', fontsize=12)
plt.tight_layout()
plt.show()

# quarter wise
plt.figure(figsize=(10, 6))
sns.countplot(x='Quarter', data=df, palette='Set2')
plt.title('Quarterly Car Sales', fontsize=14)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Number of Cars Sold', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 13.How does the car price distribution change when considering different combinations of body style and transmission type ?
plt.figure(figsize=(12, 6))
sns.boxplot(x='Body Style', y='Price ($)', hue='Transmission', data=df, palette='Set2')
plt.title('Car Price Distribution by Body Style and Transmission Type', fontsize=14)
plt.xlabel('Body Style', fontsize=12)
plt.ylabel('Car Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 15. How does the average car price vary across different car models and engine types?
# Group by car model and engine type, and calculate the average price
avg_price_by_model_engine = df.groupby(['Model', 'Engine'])['Price ($)'].mean().reset_index()

# Bar plot to visualize average price by model and engine type
plt.figure(figsize=(20, 10))
sns.barplot(x='Model', y='Price ($)', hue='Engine', data=avg_price_by_model_engine, palette='Set2')
plt.title('Average Car Price by Model and Engine Type', fontsize=14)
plt.xlabel('Car Model', fontsize=12)
plt.ylabel('Average Car Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
